$$
\bar{x} = \frac{\sum x_i}{n}
$$

```python
##Mean is a measure of central tendency that is calculated by adding up all the values in a dataset and then dividing by the number of values. It is often referred to as the "average" and is commonly used to summarize data. The mean can be sensitive to outliers, which are extreme values that can skew the average. In such cases, other measures of central tendency, such as the median or mode, may be more appropriate.

## Median is another measure of central tendency that represents the middle value in a sorted dataset. To find the median, you first arrange the values in order from smallest to largest. If there is an odd number of values, the median is the value that is exactly in the middle. If there is an even number of values, the median is calculated by taking the average of the two middle values. The median is less affected by outliers compared to the mean, making it a more robust measure of central tendency in certain situations.

## Mode is a measure of central tendency that represents the value or values that occur most frequently in a dataset. A dataset can have one mode (unimodal), more than one mode (multimodal), or no mode at all if all values occur with the same frequency. The mode is particularly useful for categorical data, where it can indicate the most common category or value. In numerical data, the mode can provide insights into the most frequently occurring value, but it may not always be a good representation of the central tendency, especially if there are multiple modes or if the data is continuous.
```

```python
## Program of mean
def calculate_mean(data):
    if len(data) == 0:
        return None  # Return None for an empty dataset
    total = sum(data)
    mean = total / len(data)
    return mean 
```

```python
## Program of median
def calculate_median(data):
    if len(data) == 0:
        return None  # Return None for an empty dataset
    sorted_data = sorted(data)
    n = len(sorted_data)
    mid = n // 2
    if n % 2 == 0:  # Even number of values
        median = (sorted_data[mid - 1] + sorted_data[mid]) / 2
    else:  # Odd number of values
        median = sorted_data[mid]
    return median
```

```python
## Program of mode
from collections import Counter
def calculate_mode(data):
    if len(data) == 0:
        return None  # Return None for an empty dataset
    data_count = Counter(data)
    max_count = max(data_count.values())
    modes = [key for key, count in data_count.items() if count == max_count]
    if len(modes) == len(data_count):
        return None  # No mode if all values occur with the same frequency
    return modes
```
