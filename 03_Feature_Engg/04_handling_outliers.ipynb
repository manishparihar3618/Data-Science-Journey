## Handling Outliers Using Python

### 5 number Summary and Box Plot

```python
## Minimum, Maximum, Median, Q1, Q3, IQR
```

```python
import numpy as np 
```

```python
lst_marks = [45,32,56,76,87,75,32,89,98,73,45,66,34,55,44]
#Calculate quartiles for outlier detection
minimum,Q1,median,Q3,maximum = np.quantile(lst_marks,[0,0.25,0.50,0.75,1.0]) 
```

```python
minimum,Q1,median,Q3,maximum
```

```python
#ISR measures the spread of the middle 50% of data
IQR=Q3-Q1
IQR
```

```python
#Values outside these limits are treated as outliers
lower_fence = Q1-1.5*(IQR)
higher_fence = Q3+1.5*(IQR)
```

```python
lower_fence
```

```python
higher_fence
```

```python
import seaborn as sns
```

```python
sns.boxplot(lst_marks) # No outliers detected
```

Outliers

```python
lst_marks = [-100,-200,45,32,56,76,87,75,32,89,98,73,45,66,34,55,44,1200,1800] #extreme value added to demonstrate outlier detection
```

```python
sns.boxplot(lst_marks) #Points are potential outliers
```

```python
# Find outliers using ISR 
outliers = [x for x in lst_marks if x < lower_fence or x > higher_fence]
print(outliers)
```

```python
# Remove observations outside IQR boundaries
clean_data = [x for x in lst_marks
              if lower_fence <= x <= higher_fence]

print(clean_data)
```

```python
sns.boxplot(clean_data)
```

## REVISION:
Q1 = 25th Percentile

Q3 = 75th Percentile

IQR = Q3 - Q1

Lower Fence = Q1 - 1.5 * IQR

Upper Fence = Q3 + 1.5 * IQR

Values outside fences are Outliers

Boxplot is the quickest way to detect outliers visually

Handling Method used = Removal
### Outlier = Observation significantly different from the rest of the data
